### Making projections and sinograms from a random image:

```python
#libs
import numpy as np
import skimage as sk
import imageio as io
import matplotlib.pyplot as plt

from skimage.transform import rotate
from skimage.color import rgb2gray
```

```python
#read shapes.tiff
image = io.imread('C:/Users/Matthew/Desktop/shapes.tiff')
#if RGB uncomment below
image = image[:,:,1]
plt.imshow(image, cmap = 'gray')
plt.title('Image')
plt.show()
```

```python
image_pad = np.pad(image, 200)
plt.imshow(image_pad, cmap = 'gray')
plt.title('Image wt padds')
plt.show()
```

```python
projs = []
for angle in range(360):
    img = rotate(image_pad, angle)
    projs.append(img)
    
plt.imshow(projs[90], cmap = 'gray')
plt.title('Image Rotated 90')
plt.show()
```

```python
projs = np.array(projs)
```

```python
# now that we have 'projections' we can plot a sinogram by reslicing along the other dimension
plt.imshow(projs[:,500])
plt.title('Sinogram')
plt.ylabel('angles')
plt.xlabel('x-axis')
plt.show()
```
